```python
#scarping the webpage
import re
import requests
import nltk 
from bs4 import BeautifulSoup
from collections import Counter
from nltk.tokenize import sent_tokenize
```

```python
#scraping webpage... getting html
url = requests.get('https://www.azlyrics.com/lyrics/akon/beautiful.html')
url.encoding = 'utf-8'
html = url.text

#extracing all tags from the html 
soup  = BeautifulSoup(html)
text  = soup.text

#removing all white spaces and using only words
tokenizer = nltk.tokenize.RegexpTokenizer('\w+')
token     = tokenizer.tokenize(text)
clean_text = str(re.sub("\s{2,}"," ",text))

#saving data to file.. changing first to string
with open('raw_speech.txt', 'w', encoding='utf-8') as f:
    f.write(clean_text)

#reading from the file
with open('raw_speech.txt', 'r', encoding='utf-8') as f:
    print(f.read())
```
